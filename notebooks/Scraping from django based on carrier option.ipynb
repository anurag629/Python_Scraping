```python
# importing two modules named selenium and pandas
# pandas for converting scraped data into exel file
# selenium for scraping data

import pandas as pd
from selenium import webdriver
```

```python
# run the chrome driver .exe file and then paste link of site from where you want to scrape the data
# I have selected the arts and design and pasted the link 

browser = webdriver.Chrome("chromedriver.exe")
browser.get("https://www.linkedin.com/jobs/arts-and-design-jobs-mumbai?keywords=Arts%20And%20Design&location=Mumbai%2C%20Maharashtra%2C%20India&locationId=&geoId=106164952&sortBy=R&f_TPR=&distance=100&position=1&pageNum=0")
```

```python
# scraping data by using class of html
# after scraping data use for loop to collect text of the data and store it in list.
# after storing in list print and check wheather you have got correct details or not

job_position = browser.find_elements_by_class_name("base-search-card__title")
company_offering = browser.find_elements_by_class_name("base-search-card__subtitle")
location_job = browser.find_elements_by_class_name("job-search-card__location")
job_position_l = []
company_offering_l = []
location_job_l = []

for i in job_position:
    job_position_l.append(i.text)
    
for i in company_offering:
    company_offering_l.append(i.text)
    
for i in location_job:
    location_job_l.append(i.text)
      
print(job_position_l)
print(len(job_position_l))

print(company_offering_l)
print(len(company_offering_l))

print(location_job_l)
print(len(location_job_l))
```

```python
# here we have created dataframe named 'df' and store vales in it.

df = pd.DataFrame({"Job Position":job_position_l, "Company offering":company_offering_l, "Location":location_job_l})
df
```

```python
# here we have converted data frame into excel file.

df.to_excel("carrier option to linkdln.xlsx",sheet_name="carrier option linkdln")
```
