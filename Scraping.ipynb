```python
# importing two modules named selenium and pandas
# pandas for converting scraped data into exel file
# selenium for scraping data

import pandas as pd
from selenium import webdriver
```

```python
# run the chrome driver .exe file and then paste link of site from where you want to scrape the data

browser = webdriver.Chrome("chromedriver.exe")
browser.get("https://www.careerguide.com/career-options")
```

```python
# scraping data by using class of html
# after scraping data use for loop to collect text of the data and store it in list named 'c'
# after storing in list print and check wheather you have got correct details or not

job_types = browser.find_elements_by_class_name("c-content-list-1")
c = []
for i in job_types:
    c.append(i.text)
    
print(c)
print(len(c))
```

```python
# we use another list to store the data into correct formt

c_job_types = []
for i in c:
    c_job_types.append((i.split('\n')))
print(c_job_types)    
print()
print(len(c_job_types))
```

```python
# above we have scraped types of job now we scrape the title of job

job_types_title = browser.find_elements_by_class_name("c-font-bold")
cc = []
for i in job_types_title:
    cc.append(i.text)
print(cc)
print(len(cc))
```

```python
#above we have got some empty string, so we have to remove them

col = cc[1:40]
print(col)
print(len(col))
```

```python
#now we store title of job and there types into dictionary, so we can convert store them in exel file

res = {col[i]: c_job_types[i] for i in range(len(col))}
print(res)
print(len(res))
```

```python
# here we have created dataframe named 'df' and store vales in it.

df = pd.DataFrame({ key:pd.Series(value) for key, value in res.items() })
df
```

```python
# here we have converted data frame into excel file.

df.to_excel("carrier option.xlsx",sheet_name="carrier option job types")
```
